```python
while i < 500:
    url = "https://www.imdb.com/search/title/?title_type=tv_series&countries=in&start={}&ref_=adv_nxt".format(i);
    req = scraper.get(url)
    webp = req.text
    soup = BeautifulSoup(webp, 'html.parser')
    all_headers = soup.findAll("h3","lister-item-header")
    for header in all_headers:
        link = header.a['href']
        link = 'https://www.imdb.com' + link
        links.append(link)
    i+=50
```

```python
def name(soup):
    return soup.find('h1').text.strip()
```

```python
# Duration
def duration(soup):
    try:
        return soup.find('time').text.strip()
    except:
        return np.nan
```

```python
# Genre
def genre(soup):
    try:
        tags = soup.find('div','subtext').findAll('a')[:-1]
        genres = []
        for t in tags:
            genres.append(t.text.strip())
        return genres
    except:
        return np.nan
```

```python
# Ratings
def ratings(soup):
    try:
        return float(soup.find('div','imdbRating').div.strong.text)
    except:
        return np.nan
```

```python
# No. of Ratings
def no_ratings(soup):
    try:
        return int(soup.find('div','imdbRating').find('a').text.replace(',',''))
    except:
        return np.nan
```

```python
# No. of Episodes
def episodes(soup):
    try:
        return int(soup.find('div','button_panel navigation_panel').span.text[:-8].strip())
    except:
        return np.nan
```

```python
# Creators 
def creators(soup):
    try:
        creators = []
        if len(soup.findAll('div','credit_summary_item')) == 1:
            return np.nan
        else:
            for creator in soup.findAll('div','credit_summary_item')[0].findAll('a'):
                creator = creator.text
                creators.append(creator)
        return creators
    
    except:
        return np.nan
```

```python
# Stars
def stars(soup):
    try:
        stars = []
        idx = 1
        if len(soup.findAll('div','credit_summary_item')) == 1:
            idx = 0
        for star in soup.findAll('div','credit_summary_item')[idx].findAll('a'):
            star = star.text
            stars.append(star)
        stars = stars[:-1]
        return stars
    except:
        return np.nan
```

```python
# Reviews (Users)
def reviews_users(soup):
    try:
        return int(soup.find('span','subText').findAll('a')[0].text[:-4].replace(',','').strip())
    except:
        return np.nan
```

```python
# Reviews (Critics)
def reviews_critics(soup):
    try:
        return int(soup.find('span','subText').findAll('a')[1].text[:-6].replace(',','').strip())
    except:
        np.nan
```

```python
# Seasons 
def seasons(soup):
    try:
        return int(soup.find('div','seasons-and-year-nav').findAll('a')[0].text.strip())
    except:
        return np.nan
```

```python
# Storyline
def storyline(soup):
    try:
        return soup.find('div','inline canwrap').text.strip()
    except:
        return np.nan
```

```python
# Language
def language(soup):
    try:
        return soup.find('h4',text='Language:').parent.a.text
    except:
        return np.nan
```

```python
# Release Date 
def release_data(soup):
    try:
        return soup.find('h4',text='Release Date:').parent.text.replace("Release Date:","").replace("    \nSee more\xa0»\n    \n","").strip()
    except:
        return np.nan
```

```python
def row_maker(soup):
    row = []
    row.append(name(soup))
    row.append(duration(soup))
    row.append(genre(soup))
    row.append(ratings(soup))
    row.append(no_ratings(soup))
    row.append(episodes(soup))
    row.append(creators(soup))
    row.append(stars(soup))
    row.append(reviews_users(soup))
    row.append(reviews_critics(soup))
    row.append(seasons(soup))
    row.append(storyline(soup))
    row.append(language(soup))
    row.append(release_data(soup))
    return row
```

```python
def table_maker(links_list):
    table = []
    count = 0
    for i in links_list:
        req = scraper.get(i)
        webp = req.text
        soup = BeautifulSoup(webp, 'html.parser')
        print(i)
        print(count)
        table.append(row_maker(soup))
        count+=1
        
    tab_arr = np.array(table)
    return tab_arr
```

```python
top_500 = table_maker(links)
```

```python
# Scraping required data
try:
    html = requests.get('https://www.imdb.com/chart/top/')
    html.raise_for_status()

    soup = BeautifulSoup(html.text,'html.parser')
    movies = soup.find('tbody',class_='lister-list').find_all('tr')
    for movie in movies:
        name = movie.find('td',class_='titleColumn').a.text
        rank = movie.find('td',class_='titleColumn').get_text(strip=True).split('.')[0]
        year = movie.find('td',class_='titleColumn').span.text.strip('()')
        rating = movie.find('td',class_='ratingColumn imdbRating').strong.text
        print(rank,name, year, rating)
        excel_sheet.append([rank,name, year, rating])
except Exception as e:
    print(e)
```
